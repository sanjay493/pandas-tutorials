```python
df["Mineral"]=df["Mineral"].str.strip().str.upper()
```

```python
df["Mineral"]
```

```python
df["Mineral"]=df["Mineral"].str.replace("0","O",regex=False)
```

```python
df["Mineral"]
```

```python
df["Mineral"]=df["Mineral"].str.replace("QUARTZ","QUARTZITE",regex=False)
```

```python
df["Mineral"]
```

```python
df["Mineral"]=df["Mineral"].str.replace("QUARTZITEITE","QUARTZITE",regex=False)
```

```python
df["Mineral"].value_counts()
```

```python
year_cols = df.columns.drop("Mineral")
year_cols
```

```python
df[year_cols] = df[year_cols].apply(pd.to_numeric, errors="coerce")
```

```python
grouped_df = df.groupby("Mineral")[year_cols].sum()
grouped_df
```

```python
grouped_df.describe()
```

```python
grouped_df=grouped_df[(grouped_df[year_cols]!=0).any(axis=1)]
```

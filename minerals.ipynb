```python
import pandas as pd
```

```python
df=pd.read_csv('mineral_wise_revenue_collected.csv')
df.info()
```

```python
df
```

```python
df.loc[:, df.columns != "Mineral"] = df.loc[:, df.columns != "Mineral"].apply(
    pd.to_numeric, errors="coerce"
)
```

```python
df
```

```python
col_values=df.columns[df.columns !="Mineral"]
```

```python
df=df[df[col_values].sum(axis=1)!=0]
```

```python
df
```

```python
df.loc[:, "Mineral"] = df["Mineral"].replace("C0AL", "Coal")
df.loc[:, "Mineral"] = df["Mineral"].replace("Quartzite", "Quartz")
df
```

```python
df["Mineral"].replace("Quartzite","Quartz")
```

```python
years_cols = df.columns[df.columns!="Mineral"]
years_cols
```

```python
df.groupby("Mineral", as_index=False)[years_cols].sum()
```

```python
df.to_csv("Modified_Minerals.csv")
```
